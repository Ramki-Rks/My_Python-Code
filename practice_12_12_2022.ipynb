```python
from itertools import combinations
```

```python
def power_set(dict_):
  output=[]
  for i in range (0,4):
    subsets=combinations(dict_,i)
    for j in subsets:
      output.append(set(j))

  return print(output)
```

```python
dict_={1,2,3}
power_set(dict_)
```
