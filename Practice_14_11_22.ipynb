```python
def practice(list_):

  output=[]
  j=0
  for i in list_:
    if i ==list_[0]:
      output.append(i)
    elif i>list_[j]:
      output.append(i)
      j+=1
    else:
      pass
  
  return sum(output)


list_ = [2,4,6,2,5]
practice(list_)
```
