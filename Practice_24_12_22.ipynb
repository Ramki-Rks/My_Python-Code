```python
def maximum_sum(array):
  output=[]

  for i in array:
    position=array.index(i)
    subarray=array[position+1 : len(array)+1]
    output.append(sum(subarray))

  return max(output)
```

```python
array=[34,-50,42,14,-5,86]
maximum_sum(array)
```
