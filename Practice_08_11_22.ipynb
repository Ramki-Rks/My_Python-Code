```python
def product_num_except_index(list_):
  output=[]
  lists=list_
  for i in lists:
    initial=1
    num=list_
    num.remove(i)
    print((num))
 
    for j in num:
      initial=initial *j
    output.append(initial)
    num.append(i)
    initial=1


  return sorted(output,reverse=True)


list_=[1,2,3,4,5]
product_num_except_index(list_)
```
