```python
def non_duplicate(list_):
  output={}
  for i in list_:
    if i in output.keys():
      output[i]+=1
    else:
      output[i]=1
  for j,k in output.items():
    if k==1:
      return j
    else:
      pass


  return output
```

```python
list_=[6,1,3,3,3,6,6]
non_duplicate(list_)
```
