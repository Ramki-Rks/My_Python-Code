```python
def autocomplete_system(query, strings):
  output=set()
  for i in strings:
   if query in i:
     output.add(i)
   else:
      pass
  
  return output


query=input("Enter the query string: ")
strings=["dog","deer","deal","deer"]
autocomplete_system(query,strings)
```
