```python
import pandas as pd
```

```python
def encoding_string(sample):
  d = pd.Series(list(sample),index=[i for i in range(len(sample))])

  charac_count=[(cat[1], grp.shape[0]) for cat, grp in d.groupby([d.ne(d.shift()).cumsum(), d])]
  
  output=[]
  for i,j in charac_count:
    output.append(f'{j}{i}')

  #combined_string="".join(output)

  return ("".join(output))

```

```python
strings="AAAABBBCCDAA"
encoding_string(strings)
```
