```python
def median(lst):
  for i in range(0, len(lst)):

    sortedLst = sorted(lst[0:i+1])
    lstLen = len(sortedLst)
    index = (lstLen - 1) // 2
   
    if (lstLen % 2):
        print(sortedLst[int(index)])
    else:
        print((sortedLst[int(index)] + sortedLst[int(index + 1)])/2.0)
```
